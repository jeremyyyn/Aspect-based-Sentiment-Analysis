```python
import pandas as pd

df = pd.read_csv(f"{gdrive_path}Preprocessed_Scraped_reviews.csv", on_bad_lines='skip')
df.sample(n=10)
```

```python
#Prepare data for training
from ast import literal_eval

data_words = []
for x in df['Clean']:
    data_words.append(' '.join(literal_eval(x)))
```

```python
#Train BERTopic
from bertopic import BERTopic

#topic_model = BERTopic(language="english", calculate_probabilities=True, verbose=True)
topic_model = BERTopic(language="english", verbose=True)
topics, probs = topic_model.fit_transform(data_words)
```

```python
import pickle

topic_model.get_topic_info().head(100).to_excel(f'{gdrive_path}BERT_topics.xlsx')
pickle.dump(topic_model, open(f'{gdrive_path}BERTopic_model', 'wb'))
topic_model.save(f'{gdrive_path}BERTopic_model')
```
