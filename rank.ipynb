```python
import pandas as pd

rank_list = pd.read_csv('labelled_aspect_3May23.csv')
rank_list = rank_list.dropna().drop(columns=rank_list.columns[0:2])
rank_list = rank_list.drop_duplicates()
print(rank_list['product'].nunique())
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

model = keras.models.load_model('sg_Sentiment_lstm_w2v.sav')
```

```python
X = rank_list['review'].values
y = []
```

```python
import pickle
#pickle.load()
token_load = open('sg_lstm_w2v_tokenizer.pkl', "rb")
tokenizer = pickle.load(token_load)
```

```python
def decode_sentiment(score, include_neutral=False):
    if include_neutral:        
        label = NEUTRAL
        if score <= SENTIMENT_THRESHOLDS[0]:
            label = NEGATIVE
        elif score >= SENTIMENT_THRESHOLDS[1]:
            label = POSITIVE

        return label
    else:
        return 0 if score < 0.5 else 1
```

```python
SEQUENCE_LENGTH = 100

# SENTIMENT
POSITIVE = "positive"
NEGATIVE = "negative"
NEUTRAL = "neutral"
SENTIMENT_THRESHOLDS = (0.4, 0.7)
```

```python
X = pad_sequences(tokenizer.texts_to_sequences(X), maxlen=SEQUENCE_LENGTH)
```

```python
[X]
```

```python
y = model.predict(X)
```

```python
y = list(map(decode_sentiment, y))
```

```python
rank_list['Label'] = y
rank_list
```

```python


positive_rank_list = rank_list[rank_list['Label'] == 1]

grouped_products = positive_rank_list.groupby('product')

sum_aspect_values = grouped_products[['Topic 0', 'Topic 1', 'Topic 2', 'Topic 3', 'Topic 4', 'Topic 5']].sum()

sorted_aspect_values = sum_aspect_values.apply(lambda x: x.sort_values(ascending=False), axis=0)

top_5_products_by_aspect = {}
for aspect in sorted_aspect_values.columns:
    top_5_products_by_aspect[aspect] = sorted_aspect_values[aspect].nlargest(5)

for aspect, top_products in top_5_products_by_aspect.items():
    print(f"Top 5 products for {aspect}:\n{top_products}\n")
```
